```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the table from the webpage
url = "https://worldpopulationreview.com/country-rankings/coffee-consumption-by-country"

tables = pd.read_html(url)
df = tables[0]
print(df.head())
print(df.columns)

df_clean = df.dropna()

top10 = df_clean.sort_values(
    by="Coffee Consumption Rate 2022 (kg/capita)↓",
    ascending=False
).head(10)

print(top10)

plt.figure(figsize=(10,6))
plt.bar(top10["Country"], top10["Coffee Consumption Rate 2022 (kg/capita)↓"])
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Countries by Coffee Consumption per Capita")
plt.ylabel("Kg per year")
plt.xlabel("Country")
plt.tight_layout()
plt.show()

df.to_csv("coffee_consumption_worldpopulationreview.csv", index=False)
```
